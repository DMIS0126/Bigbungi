```python
import pandas as pd
df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
# df.head()
f5_mean = df['f5'].mean()
f5_std = df['f5'].std()
# print(f5_mean, f5_std)
df['z'] = (df['f5'] - f5_mean) / f5_std
print(df['z'].median())
```
