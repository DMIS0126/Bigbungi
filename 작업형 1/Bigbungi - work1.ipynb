#### 2. 결측치 처리하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# RM 컬럼에 대한 결측치 처리를 평균값으로 대치 후 표준편차와 결측치를 삭제한 후 표준편차의 차이의 절댓값을 구하기
rm_mean = df['RM'].mean()
# print(rm_mean)
df1 = df.copy()
df2 = df.copy()
# print(df1, df2)
df1['RM'] = df1['RM'].fillna(rm_mean)
df2['RM'] = df2['RM'].dropna()
# print(df1['RM'].describe(), df2['RM'].describe())
sigma1 = df1['RM'].std()
sigma2 = df2['RM'].std()
print(abs(sigma1 - sigma2))
```

#### 3. 이상값 처리하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# ZN 컬럼을 대상으로 ZN 값의 평균값에서 표준편차의 1.5배보다 크거나 작은 ZN 값의 합 구하기
zn_mean = df['ZN'].mean()
zn_sigma = df['ZN'].std()
# print(zn_mean, zn_sigma)
df1 = df[(df['ZN']>zn_mean + 1.5*zn_sigma) | (df['ZN']<zn_mean - 1.5*zn_sigma)]
print(df1['ZN'].sum())
```

#### 5. 순위 구하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# MEDV 컬럼 기준 30번째로 큰 값을 1~29번쨰로 큰 값에 적용 후 MEDV의 평균, 중위값, 최솟값, 최댓값 순으로 한 줄에 출력하기
df = df.sort_values('MEDV', ascending = False)
# print(df.head(50))
# 왜 iloc로 하면 되고 loc로 하면 안될까
medv_30 = df.iloc[29,13]
df['MEDV'][:29] = medv_30
# df['MEDV'].head(30)
medv_mean = df['MEDV'].mean()
medv_median = df['MEDV'].median()
medv_min = df['MEDV'].min()
medv_max = df['MEDV'].max()
print(medv_mean, medv_median, medv_min, medv_max)
```

#### 6. 그룹별 집계 / 요약하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# TAX 컬럼이 TAX 컬럼의 중위값보다 큰 데이터를 대상으로 CHAS, RAD 칼럼 순으로 그룹을 지은 후 
# 각 그룹의 데이터값 별 개수 구하기
tax_med = df['TAX'].median()
df = df[df['TAX']>tax_med][['CHAS', 'RAD']]
# df.info()
df1 = df.groupby(['CHAS', 'RAD']).value_counts()
df2 = pd.DataFrame(df1, columns = ['COUNT'])
print(df2)
# 열 이름 지정이 안되는데 이유를 모르겠다.
# 알아냈다. pd.DataFrame 함수 사용할 때, 컴마하고 컬럼 이름을 리스트로 지정해줘야 한다.
```

#### 7. 오름차순 / 내림차순 정렬하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# TAX 컬럼을 오름차순으로 정렬한 결과와 내림차순으로 정렬한 결과를 구해 각 순번에 맞는 오름차순과 내림차순 값의 차이를 구해 분산 구하기
# 여기서 말하는 차이가 차네 절댓값
# sort_values, reset_index 알아놓기
df_asc = df.sort_values('TAX', ascending = True).reset_index(drop = True)
df_desc = df.sort_values('TAX', ascending = False).reset_index(drop = True)
# print(df_asc.head(5))
# print(df_desc.head(5))
df_diff = abs(df_desc['TAX']-df_asc['TAX'])
diff_var = df_diff.var()
print(diff_var)
```

#### 8. MinMaxScaling하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# MEDV 컬럼을 최소최대척도(MinMaxScaler)로 변환한 후, 0.5보다 큰 값을 가지는 레코드 수 구하기
medv_max = df['MEDV'].max()
medv_min = df['MEDV'].min()
df['minmax'] = (df['MEDV'] - medv_min) / (medv_max - medv_min)
df['minmax']
print(len(df[df['minmax']>0.5]))
```

#### 9. 빈도값 구하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# AGE 컬럼을 소수점 첫 번째 자리에서 반올림하고, 가장 많은 비중을 차지하는 AGE 값과 그 개수를 차례대로 출력하기
# 즉, AGE 컬럼의 최빈값과 그 개수 출력하기
df['AGE'] = round(df['AGE'], 0)
age_mode = df['AGE'].mode()
# print(age_mode)
# 이거 출력하면 0 100 이렇게 object로 떠서 그 값 사용할 때는 [0]과 같이 인덱스 지정 필요
df1 = df['AGE'] == age_mode[0]
print(int(age_mode[0]), df1.sum())
```

#### 10. 표준화하기

```python
import pandas as pd
df = pd.read_csv('boston.csv')
df.head(5)
```

```python
# DIS 컬럼을 표준한 뒤, 0.4보다 크고 0.6보다 작은 표준화값들에 대한 평균 구하기 
# 소수점 셋째 자리에서 반올림
dis_mean = df['DIS'].mean()
dis_std = df['DIS'].std()
# print(dis_mean, dis_std)
df['z'] = (df['DIS']-dis_mean) / dis_std
df1 = df[(df['z'] > 0.4) & (df['z'] < 0.6)]['z']
print(round(df1.mean(), 2))
```
